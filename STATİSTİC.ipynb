# Problem:
Web sitemizde geçirilen ortalama süre gerçekten 170 saniye mi ?

## Detaylar:
  Yazılımlardan elde edilen web sitesinde geçirilen ort. süreler var.
  Bu veriler incelendiğinde bir yönetici yada çalışanımız bu değerlerin böyle olmadığına yönelik düşünceler taşıyor ve bu durumu test etmek istiyorlar.

<d><i>H</i><sub>0</sub>: <i>μ</i> = <i>170</i></d>   
<d><i>H</i><sub>1</sub>: <i>μ</i> ≠ <i>170</i></d><br><br>

**Test İstatistiği:**

${\displaystyle t={\frac {{\bar {x}}-\mu _{0}}{\frac {s}{\sqrt {n}}}}}$


```python
import numpy as np

olcumler = np.array([17, 160, 234, 149, 145, 107, 197, 75, 201, 225, 211, 119, 
              157, 145, 127, 244, 163, 114, 145,  65, 112, 185, 202, 146,
              203, 224, 203, 114, 188, 156, 187, 154, 177, 95, 165, 50, 110, 
              216, 138, 151, 166, 135, 155, 84, 251, 173, 131, 207, 121, 120])
```

```python
olcumler[0:10]
```

```python
import scipy.stats as stats
import pandas as pd
```

```python
stats.describe(olcumler)
```

```python
# varsayımlar
# Normallik varsayımı bu varsayım sağlanırsa testimizi gerçekleştirebilriz.
```

```python
# histogram
pd.DataFrame(olcumler).plot.hist(); # Görsel olarak normal dağılım olup olmadığını kontrol ederiz
```

```python
# qqplot
import pylab
stats.probplot(olcumler, dist="norm", plot=pylab)
pylab.show() #teorik dağılım ile örnek dağılım arasında değerlendirme imkanı sunar ve normal dağılıp dağılmadığına ilişkin bilgi sahibi oluruz.
# ordered values(örnek dağılım noktaları) Theoretical quantiles a benziyorsa yani noktalar kırmızı çizgi etrafında konumlanıyorsa teorik dağılım ile örnek dağılım birbirine benzediği görülür.
# Normal varsayabiliriz.
```

```python
#Shapiro-Wilks Testi 
# amaç = elimizdeki bir sayısal değişkenin ortalamasının yani örneğimizin ortalamasının belirli bir değere eşit olup olmadığını sınamak
# Dağılım normal mi değil mi kısaca
# aşağıdaki hipotezler için yapılır bu test önemli
```

H0: Örnek dağılımı ile teorik normal dağılım arasında istatiksel olarak anlamlı bir farklılık yoktur.

H1: Örnek dağılımı ile teorik normal dağılım arasında istatiksel olarak anlamlı bir farklılık vardır. 

```python
from scipy.stats import shapiro 
```

```python
shapiro(olcumler) # pvalue < 0.05(alfa değeri) ten küçükse ho reddederiz genelde
#pvalue > 0.05(alfa değeri) den büyük ise HO ı reddemeyiz genelde
# mesela bu durumda pvalue=0.78 değeri alfa değeri= 0.05(genelde %95 güven seçildiğinden) büyük olduğu için HO ı reddedemeyiz.
# Ho: Örnek dağılımı ile teorik normal dağılım arasında istatiksel olarak anlamlı bir farklılık yoktur.
# Ho' ı reddemezsek T testini(Hipotez testi tek örneklem) uygulayabiliriz.Çünkü normal dağılım varsayımı sağlanmkatadır.
```

```python
print("T Hesap İstatistiği: " + str(shapiro(olcumler)[0]))
print("Hesaplanan P-value: " + str(shapiro(olcumler)[1]))
```

```python
# Bu durumda H0 hipotezi reddedilemez çünkü pvalue > alfa değeri(0,05) den
# buna göre göre hipotez testini uygulayabiliriz.
```

```python
# Hipotez Testinin Uygulanması 
```

```python
stats.ttest_1samp(olcumler, popmean =170)
# Pvalue değeri alfa 0.05 değerinden küçük olduğundan dolayı HO hipotezi reddedilir.
```

H0: Web sitemizde geçirilen ortalama süre 170'tir

H1: Web sitemizde geçirilen ortalama süre 170 değildir

 Pvalue değeri alfa 0.05 değerinden küçük olduğundan dolayı HO hipotezi reddedilir.

# Tek Örneklem Oran Testi

Problem:

Bir yazılım ile bir mecrada reklam verilmiş ve bu reklama ilişkin yazılım tarafından 0.125 dönüşüm oranı elde edildiği ifade edilmiş.Fakat bu durum kontrol edilmek isteniyor. Çünkü bu yüksek bir oran ve gelirler incelendiğinde örtüşmüyor.

Detaylar:

500 kişi dış mecrada reklamlara tıklamış, 40 tanesi sitemize gelip alışveriş yapmış.

Örnek üzerinden elde edilen dönüşüm oranı: 40/500 = 0.08

```python
from statsmodels.stats.proportion import proportions_ztest
```

```python
count = 40
nobs = 500
value = 0.125
```

```python
proportions_ztest(count,nobs,value) # soldaki: test istatistiği burda z test istatistiği
#sağdaki pvalue < alfa(0,.05) ten küçük HO ı reddediyoruz
#yorum: ürün satın alımına yönelik olarak test ettiğpimiz 0.125 e şit olması %95 güven payı ile reddedilir
```

## Korelasyon Varsayım Kontrolü
#### Normallik Varsayımı

H0: Örnek dağılımı ile teorik normal dağılım arasında istatiksel olarak anlamlı bir farklılık yoktur.

# Hipotez Testi
### Korelasyon Katsayısı

```python
df["tip"].corr(df["total_bill"]) # korelasyon katsayısı bulma
# ön tanımlı olarak pearson korelasyon kat sayısını verir(normallik varsayımı sağlandğında hesaplanabilir)
# yorum:(0.57666...) Değişkenlerin arasında pozitif yönlü orta şiddetli bir ilişki var 
```

```python
df["tip"].corr(df["total_bill"], method= "spearman") # spearman korelasyon kat sayısı
# yorum:(0.5936...) Değişkenlerin arasında pozitif yönlü orta şiddetli bir ilişki var 
```

```python
from scipy.stats.stats import pearsonr
```

```python
test_istatistigi, pvalue = pearsonr(df["tip"],df["total_bill"])

print('Korelasyon Katsayısı = %.4f, p-değeri = %.4f' % (test_istatistigi, pvalue))
```

yorum: Değişkenler arasında anlamlı bir ilişki yoktur diyen H0 hipotesizini reddediyoruz. Yani değişkenler arasında anlamlı bir ilişki vardır.

## Nonparametrik Hipotez Testi

```python
test_istatistigi, pvalue = stats.spearmanr(df["tip"],df["total_bill"])

print('Korelasyon Katsayısı = %.4f, p-değeri = %.4f' % (test_istatistigi, pvalue))
```

test sonucu incelendiğinde pvalue değeri < 0.05 değeri olduğu için değişkenler arasında ilişki yoktur diyen H0 hipotezi reddedilir.

```python
test_istatistigi, pvalue = stats.kendalltau(df["tip"],df["total_bill"])

print('Korelasyon Katsayısı = %.4f, p-değeri = %.4f' % (test_istatistigi, pvalue))
```
